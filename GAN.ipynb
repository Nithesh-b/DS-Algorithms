```
### Import numpy, mnist dataset or fashion_mnist dataset (your choice), dcgan.py and image_helper.py file
import numpy as np
from keras.datasets import fashion_mnist
from tensorflow.keras.optimizers import Adam
import sys  
sys.path.insert(0, '/content/sample_data')
from dcgan import DCGAN
import image_helper

(X,_),(_,_) =  fashion_mnist.load_data()
### Normalize the X
print (X)
X_train = (X - 127.5) / 127.5
print ("----------------------")
print (X_train)
X_train = np.expand_dims(X_train, axis=3)
print ("----------------------")
print (X_train)
### Call the DCGAN class object passing parameters image_shape, generator_input_dim, image_helper class object, and number of channels of mnist data 
generative_advarsial_network = DCGAN(X_train[0].shape, 100, image_helper,1)

### Train the X_train with batchsize of your choice for minimum 30000 epochs
generative_advarsial_network.train(300000, X_train, batch_size=3)
```
